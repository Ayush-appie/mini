````python
import zipfile, os

ZIP_PATH = '/content/drive/MyDrive/PneumoniaDataset/archive (7).zip'
EXTRACT  = '/content/chest_xray'

if not os.path.exists(EXTRACT):
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(EXTRACT)
    print("Dataset extracted!")
else:
    print("Already extracted.")
````

````python
import numpy as np
import cv2, random, warnings, joblib
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from tqdm import tqdm
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, confusion_matrix,
                              roc_auc_score, roc_curve, accuracy_score)
from skimage.feature import hog, local_binary_pattern
from deap import base, creator, tools, algorithms

warnings.filterwarnings('ignore')
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
print("All imports successful!")
````

````python
IMG_SIZE = (64, 64)

def load_all_images(base_path):
    X, y = [], []
    for split in ['train', 'test', 'val']:
        for label_name, label_val in [('NORMAL', 0), ('PNEUMONIA', 1)]:
            folder = Path(base_path) / split / label_name
            if not folder.exists():
                print(f"Skipping (not found): {folder}")
                continue
            imgs = (list(folder.glob('*.jpeg')) +
                    list(folder.glob('*.jpg'))  +
                    list(folder.glob('*.png')))
            print(f"Found {len(imgs)} images in {split}/{label_name}")
            for p in tqdm(imgs, desc=f"{split}/{label_name}"):
                img = cv2.imread(str(p), cv2.IMREAD_GRAYSCALE)
                if img is None:
                    continue
                X.append(cv2.resize(img, IMG_SIZE))
                y.append(label_val)
    return np.array(X), np.array(y)

X_raw, y = load_all_images('/content/chest_xray/chest_xray')
print(f"\nTotal images loaded: {len(X_raw)}")
print(f"Normal     : {(y==0).sum()}")
print(f"Pneumonia  : {(y==1).sum()}")
````

````python
def extract_features(images):
    features = []
    for img in tqdm(images, desc="Extracting HOG+LBP features"):
        # HOG features
        hog_feat = hog(img,
                       orientations=9,
                       pixels_per_cell=(8, 8),
                       cells_per_block=(2, 2),
                       feature_vector=True)
        # LBP features
        lbp = local_binary_pattern(img, P=8, R=1, method='uniform')
        lbp_hist, _ = np.histogram(lbp.ravel(), bins=10,
                                   range=(0, 10), density=True)
        features.append(np.concatenate([hog_feat, lbp_hist]))
    return np.array(features)

print("Extracting features from all 5856 images...")
X = extract_features(X_raw)
print(f"Feature vector size per image : {X.shape[1]}")
print(f"Total feature matrix shape    : {X.shape}")
````

````python
# Split into 65% train | 15% val | 20% test
X_tr, X_test, y_tr, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=SEED
)
X_train, X_val, y_train, y_val = train_test_split(
    X_tr, y_tr, test_size=0.15, stratify=y_tr, random_state=SEED
)

# Fit scaler ONLY on training data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val   = scaler.transform(X_val)
X_test  = scaler.transform(X_test)

print(f"Train : {len(X_train)} samples")
print(f"Val   : {len(X_val)} samples")
print(f"Test  : {len(X_test)} samples")
print(f"\nTrain — Normal: {(y_train==0).sum()} | Pneumonia: {(y_train==1).sum()}")
print(f"Val   — Normal: {(y_val==0).sum()}   | Pneumonia: {(y_val==1).sum()}")
print(f"Test  — Normal: {(y_test==0).sum()}  | Pneumonia: {(y_test==1).sum()}")
````

````python
KERNEL_MAP = {0: 'rbf', 1: 'linear', 2: 'poly'}

def decode_individual(ind):
    C      = 10 ** ind[0]
    gamma  = 10 ** ind[1]
    kernel = KERNEL_MAP[int(round(ind[2])) % 3]
    return C, gamma, kernel

def evaluate(individual):
    C, gamma, kernel = decode_individual(individual)
    try:
        clf = SVC(C=C, gamma=gamma, kernel=kernel,
                  random_state=SEED, class_weight='balanced')
        clf.fit(X_train, y_train)
        acc = accuracy_score(y_val, clf.predict(X_val))
        return (acc,)
    except Exception:
        return (0.0,)

creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

tb = base.Toolbox()
tb.register("gene_C",      random.uniform, -2,  4)
tb.register("gene_gamma",  random.uniform, -5,  1)
tb.register("gene_kernel", random.uniform,  0,  2)
tb.register("individual",  tools.initCycle, creator.Individual,
            (tb.gene_C, tb.gene_gamma, tb.gene_kernel), n=1)
tb.register("population",  tools.initRepeat, list, tb.individual)
tb.register("evaluate",    evaluate)
tb.register("mate",        tools.cxBlend, alpha=0.3)
tb.register("mutate",      tools.mutGaussian, mu=0, sigma=0.5, indpb=0.4)
tb.register("select",      tools.selTournament, tournsize=3)

print("Genetic Algorithm configured successfully!")
````

````python
POP_SIZE = 20
N_GEN    = 15
CXPB     = 0.6
MUTPB    = 0.3

print(f"Starting GA: {POP_SIZE} individuals x {N_GEN} generations...")
print("This will take 20-30 minutes. Please wait...")
print("-" * 60)

pop  = tb.population(n=POP_SIZE)
hof  = tools.HallOfFame(1)
stat = tools.Statistics(lambda ind: ind.fitness.values)
stat.register("max",  np.max)
stat.register("mean", np.mean)

pop, log = algorithms.eaSimple(
    pop, tb,
    cxpb=CXPB, mutpb=MUTPB, ngen=N_GEN,
    stats=stat, halloffame=hof, verbose=True
)

best_C, best_gamma, best_kernel = decode_individual(hof[0])
print("=" * 60)
print("GA COMPLETE!")
print(f"Best C      : {best_C:.6f}")
print(f"Best gamma  : {best_gamma:.8f}")
print(f"Best kernel : {best_kernel}")
print(f"Best val acc: {hof[0].fitness.values[0]:.4f}")
print("=" * 60)
```

This will print one line per generation like:
```
gen    nevals    max       mean
0      20        0.91      0.87
1      12        0.92      0.89
...
````

````python
POP_SIZE = 20
N_GEN    = 15
CXPB     = 0.6
MUTPB    = 0.3

print(f"Starting GA: {POP_SIZE} individuals x {N_GEN} generations...")
print("This will take 20-30 minutes. Please wait...")
print("-" * 60)

pop  = tb.population(n=POP_SIZE)
hof  = tools.HallOfFame(1)
stat = tools.Statistics(lambda ind: ind.fitness.values)
stat.register("max",  np.max)
stat.register("mean", np.mean)

pop, log = algorithms.eaSimple(
    pop, tb,
    cxpb=CXPB, mutpb=MUTPB, ngen=N_GEN,
    stats=stat, halloffame=hof, verbose=True
)

best_C, best_gamma, best_kernel = decode_individual(hof[0])
print("=" * 60)
print("GA COMPLETE!")
print(f"Best C      : {best_C:.6f}")
print(f"Best gamma  : {best_gamma:.8f}")
print(f"Best kernel : {best_kernel}")
print(f"Best val acc: {hof[0].fitness.values[0]:.4f}")
print("=" * 60)
````

````python
final_svm = SVC(
    C            = best_C,
    gamma        = best_gamma,
    kernel       = best_kernel,
    probability  = True,
    class_weight = 'balanced',
    random_state = SEED,
    cache_size   = 500
)

final_svm.fit(X_train, y_train)
print("Final GA+SVM model trained successfully!")
````

````python
y_pred = final_svm.predict(X_test)
y_prob = final_svm.predict_proba(X_test)[:, 1]

print("=" * 50)
print("TEST SET RESULTS")
print("=" * 50)
print(classification_report(y_test, y_pred, target_names=['NORMAL', 'PNEUMONIA']))
print(f"ROC-AUC Score : {roc_auc_score(y_test, y_prob):.4f}")
print(f"Accuracy      : {accuracy_score(y_test, y_pred):.4f}")

fig, axes = plt.subplots(1, 3, figsize=(17, 5))

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=['NORMAL', 'PNEUMONIA'],
            yticklabels=['NORMAL', 'PNEUMONIA'],
            annot_kws={"size": 14})
axes[0].set_title('Confusion Matrix', fontweight='bold')
axes[0].set_ylabel('Actual')
axes[0].set_xlabel('Predicted')

fpr, tpr, _ = roc_curve(y_test, y_prob)
auc_val = roc_auc_score(y_test, y_prob)
axes[1].plot(fpr, tpr, color='steelblue', lw=2, label=f'AUC = {auc_val:.3f}')
axes[1].plot([0,1],[0,1], 'k--', lw=1)
axes[1].fill_between(fpr, tpr, alpha=0.08, color='steelblue')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].set_title('ROC Curve', fontweight='bold')
axes[1].legend(loc='lower right')

gen_max  = [g['max']  for g in log]
gen_mean = [g['mean'] for g in log]
axes[2].plot(gen_max,  color='crimson',    lw=2, marker='o', label='Best')
axes[2].plot(gen_mean, color='darkorange', lw=2, linestyle='--', label='Mean')
axes[2].set_xlabel('Generation')
axes[2].set_ylabel('Validation Accuracy')
axes[2].set_title('GA Fitness Evolution', fontweight='bold')
axes[2].legend()

plt.suptitle('GA + SVM — Pediatric Pneumonia Detection', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('/content/results.png', dpi=150, bbox_inches='tight')
plt.show()
````
